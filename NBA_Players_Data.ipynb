```python
import pandas as pd
import numpy as np
import requests
import time
```

```python
test_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season=2023-24&SeasonType=Regular%20Season&StatCategory=PTS'
```

```python
r = requests.get(url=test_url).json()
```

```python
row_headers = r['resultSet']['headers']
```

```python
df1 = pd.DataFrame(r['resultSet']['rowSet'], columns = row_headers)
```

```python
df2 = pd.DataFrame({'Year': ['2023-24' for i in range(len(df1))],
                    'Season_Type': ['RegularSeason' for i in range(len(df1))]})
df3 = pd.concat([df2, df1], axis=1)
```

```python
years = ['2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19','2019-20','2020-21','2021-22','2022-23']
season = ['Regular%20Season', 'Playoffs']
df = pd.DataFrame(columns = ['Year', 'Season_Type'] + row_headers)

for y in years:
    for s in season:
        api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season=' + y + '&SeasonType=' + s + '&StatCategory=PTS'
        r = requests.get(url=api_url).json()
        row_headers = r['resultSet']['headers']
        df1 = pd.DataFrame(r['resultSet']['rowSet'], columns = row_headers)
        df2 = pd.DataFrame({'Year': [y for i in range(len(df1))],
                            'Season_Type': [s for i in range(len(df1))]})
        df3 = pd.concat([df2, df1], axis = 1)
        df = pd.concat([df, df3], axis = 0)
        print(f'Done scaping player data for {y} {s}')
        time_btw = np.random.uniform(low=5, high=40)
        print(f'waiting for {time_btw} seconds to start scaping')
        time.sleep(time_btw)

print(f'Scaping Completed')
df.to_excel('NBA_Data.xlsx', index = False)
```
